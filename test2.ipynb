```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
data=pd.read_csv("Book2.csv")#Read the data from CSV file
```

# Answer 20

```python
plt.bar(data['Name'], data['Department']) #plotting the bar graph of the employees and their salary(19)
plt.title('Bar graph of Name of employees in each department')
plt.xlabel('Name of employee')
plt.ylabel('Department')
```

# Answer 21

```python
salary_filtering = data[data['Salary'] > 60000] 
print(salary_filtering) #filtering the employees with salary greater than 60000
```

# Answer 22

```python
# Create a line graph for the salary trend by department
plt.figure(figsize=(10,6))
plt.plot(data.groupby("Department")["Salary"].mean())
plt.xlabel("Department")
plt.ylabel("AverageSalary")
```
